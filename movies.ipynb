## Kategorik Değişkenlerin İncelenmesi

### Original Language kategorisinde en çok tekrar eden değerler:

```python
from collections import Counter
Counter(df["original_language"]).most_common(10)
```

```python
plt.figure(figsize=(10,8))
df["original_language"].value_counts().head(10).plot(kind='barh')
plt.title("Most Used 10 Languages", size=15)
plt.show()
```

### Genre kategorisinde en çok tekrar eden değerler:

```python
Genre_Top_10=Counter(df["genre"]).most_common(10)
Genre_Top_10
```

```python
Genre_Top_10_graph=[]
value=[]
for i in Genre_Top_10:
    Genre_Top_10_graph.append(i[0])
    value.append(i[1])
```

```python
plt.figure(figsize= (10, 10))
colors = ["#FC5A50", "#FC5A50", "#7BC8F6", "#7BC8F6","#7BC8F6", "#7BC8F6", "#7BC8F6", "#7BC8F6", "#ADD8E6", "#ADD8E6"]
plt.pie(value, colors=colors, labels = Genre_Top_10_graph, textprops={'fontsize': 14},explode = [0.05,0.05,0,0,0,0,0,0,0,0], startangle = 20,shadow=True,autopct="%.2f")
plt.title ("MOVIE GENRE TOP 10", fontsize=18)
plt.show()
```
